```python
import numpy as np
import matplotlib.pyplot as plt #绘图方法
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([3,5,7,6,2,7,10,15])
plt.plot(x,y,'g') #绘制折线图 参数1：x 2：y 3：color 4.线条宽度
plt.plot(x,y,'g',lw=1)
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([13,25,17,36,21,16,10,15])
plt.bar(x,y,3,alpha=0.6,color='b')#绘制柱状图 参数1：x 2：y 3：宽度比例 4：透明度0-1 5：颜色 
plt.show()
```
