```python
#pos 888 neg 903
#1.参数设置 2.创建hog 3.创建svm 4.计算hog 5.给标签 6.训练 7.预测 8.绘图
import cv2
import numpy as np
import matplotlib.pyplot as plt

#1.参数设置
posNum = 888
negNum = 903
winSize = (64,128)
blockSize = (16,16)#105个block
blockStep = (8,8)
cellSize = (8,8)
numBin = 9

#2.创建hog
hog = cv2.HOGDescriptor(winSize,blockSize,blockStep,cellSize,numBin)#参数win,blocksize,blockstep,cellsize,numbin

#3.创建svm
svm = cv2.ml.SVM_create()
svm.setType(cv2.ml.SVM_C_SVC)
svm.setKernel(cv2.ml.SVM_LINEAR)
svm.setC(0.01)

#4.计算Hog,5.给标签
featureNum = int(3780)#特征维度:3780维
featureArray = np.zeros((posNum+negNum,featureNum),np.float32)#对每张图创建一个3780个数据的数组
labelArray =  np.zeros(((posNum+negNum),1),np.int32)

for i in range(0,posNum):
    fileName = 'C:\\Users\TH\Desktop\DL\\pos\image'+str(i+1)+'.jpg'
    img = cv2.imread(fileName,1)
    hist = hog.compute(img,(8,8))#计算结果也3780维
    for j in range(0,featureNum):
        featureArray[i,j] = hist[j] #将hog特征装入数组中
    labelArray[i,0] = 1 #正样本标签为1
    
for i in range(0,negNum):
    fileName = 'C:\\Users\TH\Desktop\DL\\neg\image'+str(i+1)+'.jpg'
    img = cv2.imread(fileName,1)
    hist = hog.compute(img,(8,8))#计算结果也3780维
    for j in range(0,featureNum):
        featureArray[i+posNum,j] = hist[j] #将hog特征装入数组中
    labelArray[i+posNum,0] = -1 #负样本标签为-1
    
#6.训练    
result = svm.train(featureArray,cv2.ml.ROW_SAMPLE,labelArray)

#7.检测 核心：创建myHog，myHog来源于myDetect,myDetect又来源于resultArray和rho
#rho用于判决，来源于svm的get，svm来源于训练
#resultArray来源于 -1*alphaArray*supportVArray
alpha = np.zeros((1),np.float32)
rho = svm.getDecisionFunction(0,alpha)
print(alpha)
print(rho)
alphaArray = np.zeros((1,1),np.float32)
supportVArray = np.zeros((1,featureNum),np.float32)#支持向量机数组
resultArray = np.zeros((1,featureNum),np.float32)
alphaArray[0,0] = alpha
resultArray = -1*alphaArray*supportVArray

print(resultArray)
myDetect = np.zeros((3781),np.float32)
for i in range(0,3780):
    myDetect[i] = resultArray[0,i]
myDetect[3780] = rho[0]


myHog = cv2.HOGDescriptor()#构建hog
myHog.setSVMDetector(myDetect)


imageSrc = cv2.imread('image1.jpg',1)
objs = myHog.detectMultiScale(imageSrc,0,(8,8),(32,32),1.05,2)#，objs三维数组，xywh放在最后一维
#(8,8)win滑动步长，（32,32）窗体大小，1.05缩放系数
x = int(objs[0][0][0])
y = int(objs[0][0][1])
w = int(objs[0][0][2])
h = int(objs[0][0][3])
cv2.rectangle(imageSrc,(x,y),(x+w,y+h),(0,0,255),2)
cv2.imshow('dst',imageSrc)
cv2.waitKey(0)
print('end!')
```
